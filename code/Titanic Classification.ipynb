# 泰坦尼克号预测问题
泰坦尼克号生存分类经典又兼具备趣味性的Kaggle案例[泰坦尼克号存活预测问题](https://www.kaggle.com/c/titanic)。
比赛说明：泰坦尼克号的沉没是历史上最臭名昭着的沉船之一。1912年4月15日，在首航期间，泰坦尼克号撞上一座冰山后沉没，2224名乘客和机组人员中有1502人遇难。这一耸人听闻的悲剧震撼了国际社会，导致了更好的船舶安全条例。
沉船导致生命损失的原因之一是乘客和船员没有足够的救生艇。虽然幸存下来的运气有一些因素，但有些人比其他人更有可能生存，比如妇女，儿童和上层阶级。
在这个挑战中，我们要求你完成对什么样的人可能生存的分析。特别是，我们要求你运用机器学习的工具来预测哪些乘客幸存下来的悲剧。

## 前言

     * 要重视对数据的认识
     * 数据中的特殊点/离群点的分析和处理很重要
     * 特征工程在很多Kaggle的场景下，甚至比模型本身还重要
     * 最后要做模型融合

## Step2 特征工程(feature engineering)
<font color=blue>对数据进行预处理，为后面的建模做准备。预处理主要包括对缺失值的补全和将object类型数据进行one-hot编码。
遇到缺失值的情况，有几种常见的处理方式：<font>
     * 如果缺失的样本占总数比例极高，可能就舍弃这个特征了。作为特征加入反而可
       能引入噪声，影响最后的结果。
     * 若缺失的样本适中，且该属性非连续值属性，可以把NaN作为一个新类别，加到类          别特征中
     * 若缺失样本适中，且该属性为连续值属性，有时候考虑给定一个步长，把它离散          化，再如上把NaN作为一个特征值加入
     * 缺失样本不多的情况下，也可以根据现有的值，拟合以下数据，补充。
     * 缺失样本不多的情况下可以补充均值或者中值

### 处理缺失值
<font color=blue>这里尝试用scikit-learn中的RandomForest来拟合缺失的年龄数据。RandomForest是一个用在原始数据中做不同采样，建立多颗Decision Tree，再进行average等等来降低过拟合现象，提高结果的机器学习算法。

```python
#补充缺失值
all_df.isnull().sum().sort_values(ascending=False).head()
```

```python
#用均值补充Fare
Fare_mean=all_df.Fare.mean()
all_df.Fare=all_df.Fare.fillna(Fare_mean)
```

```python
from sklearn.ensemble import RandomForestRegressor

#使用 RandomForestClassifier 填补缺失的年龄属性

# 把已有的数值型特征取出来丢进Random Forest Regressor中
age_df = X_train[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]

# 乘客分成已知年龄和未知年龄两部分
known_age = age_df[age_df.Age.notnull()].as_matrix()
unknown_age = age_df[age_df.Age.isnull()].as_matrix()

# y即目标年龄
y = known_age[:, 0]

# X即特征属性值
X = known_age[:, 1:]

# fit到RandomForestRegressor之中
rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
rfr.fit(X, y)
# 用得到的模型进行未知年龄结果预测
predictedAges = rfr.predict(unknown_age[:, 1::])
# 用得到的预测结果填补原缺失数据
X_train.loc[ (X_train.Age.isnull()), 'Age' ] = predictedAges 
all_df = pd.concat((X_train, X_test), axis=0)
```

```python
mean_column1=all_df.mean()
```

```python
all_df=X_train.fillna(mean_column1)
```

```python
print all_df['Embarked'].value_counts()
```

```python
#对于Embarked这种类型的特征，我们使用出现频率最高的特征值来填充，这是相对可以减少引入误差的一种方法
all_df['Embarked'].fillna('S', inplace=True)
all_df.isnull().sum().sum()
```

### 变换某些数值型变量的形式，独热码编码
<font color=blue>建立逻辑回归建模时，需要输入的特征都是数值型特征，所以在建模前要对object(类目型)的特征因子化(one-hot编码)
以Embarked为例，原本是一个特征维度，取值是['S','C','Q'],这样做逻辑回归时，模型会比较S，C，Q的ASC码的大小，这毫无意义。所以将其展开为'Embarked_C','Embarked_S','Embarked_Q'三个特征。可以使用pandas的'get_dummies'来完成这项任务。

```python
#pd.get_dummies(X_train['Cabin'], prefix='Cabin').head()
```

```python
all_df = pd.get_dummies(all_df)
dummied_Pclass=pd.get_dummies(all_df['Pclass'],prefix='Pclass')
all_df.drop(['Pclass'],axis=1,inplace=True)
all_df=pd.concat([all_df,dummied_Pclass],axis=1)
all_df.head()
```

### 归一化scaling
<font color=blue>观察Age和Fare两个属性，乘客的数值幅度变化太大。了解逻辑回归和梯度下降，可以知道各属性值之间scale差距太大，将对收敛速度造成极大减缓，甚至不收敛。所以先用scikit_learn中的preprocessing模块对这两个特征进行一个scaling，其实就是将一些变化幅度较大的特征化到[-1,1]

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
#通过行号进行索引
X_train_dummy = all_df.loc[X_train.index]
X_test_dummy = all_df.loc[X_test.index]
X_train_dummy=ss.fit_transform(X_train_dummy)            
X_test_dummy=ss.fit_transform(X_test_dummy)
X_train_dummy
```

## 建立模型，进行预测

```python
X_train_dummy.shape, X_test_dummy.shape
```

### Ridge Regression
用Ridge Regression模型来跑一遍看看。（对于多因子的数据集，这种模型可以方便的把所有的var都无脑的放进去）

```python
from sklearn.linear_model import Ridge
#cross_val_score进行交叉验证，返回值是在不同交叉情况下的准确率
from sklearn.model_selection import cross_val_score
```

```python
#用sklearn自带的cross_validation进行验证
#logspcae表示从10^-3到10^2之间取50个数
alphas = np.logspace(-3, 2, 50)
test_scores = []
for alpha in alphas:
    clf = Ridge(alpha)
    test_score = np.sqrt(-cross_val_score(clf, X_train_dummy, Y_train, cv=10, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
#存下所有的CV值，看看哪个alpha值更好（也就是『调参数』）
plt.plot(alphas, test_scores)
plt.title("Alpha vs CV Error");
```

### XGbbot模型预测

```python
from xgboost import XGBClassifier

params = range(1,10,1)
test_scores = []
score_max=0
for param in params:
    clf = XGBClassifier(max_depth=param)
    test_score = cross_val_score(clf, X_train_dummy, Y_train, cv=10, scoring='accuracy')
    test_scores.append(np.mean(test_score))
    if test_score.mean() > score_max:
        score_max=test_score.mean()
print score_max
```

```python
plt.plot(params, test_scores)
plt.title("max_depth vs CV Error");
```

```python
clf = XGBClassifier(max_depth=5)
clf.fit(X_train_dummy, Y_train)
Y_predict = clf.predict(X_test_dummy)
submission_df = pd.DataFrame(data= {'PassengerId' : test['PassengerId'], 'Survived': Y_predict})
```

## 判定当前模型所处状态(欠拟合or过拟合)
* 有一个很可能发生的问题是，我们不断地做feature engineering，产生的特征越来越多，用这些特征去训练模型，会对我们的训练集拟合地越来越好，但同时也在逐渐丧失泛化能力，从而在待预测的数据上表现不佳，也就是拟合问题。
* 欠拟合就是模型在待预测的数据上表现不佳，有可能是欠拟合。
在机器学习的问题上，对于过拟合和欠拟合的问题优化方式不同。
对过拟合而言，通常以下策略对结果优化是有用的：
* 做feature selection，挑选出较好的feature的subset来做training
* 提供更多的数据，从而弥补原始数据的bias问题，学习到的model也就更加准确。
对欠拟合，通常需要更多的feature，更复杂的模型来提高准确度
通过learning curve可以帮我们判断现在所处的状态。以样本数为横坐标，训练和交叉验证集上的错误率作为纵坐标，两种状态分别如下图所示：过拟合(overfitting/high variace),欠拟合(underfitting/high bias)
